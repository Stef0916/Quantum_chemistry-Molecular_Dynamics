```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.tri as mtri

# Define a function to read data from CSV file
def read_data_from_csv(file_path):
    data = pd.read_csv(file_path)
    alpha = data['Alpha']
    theta = data['Theta']
    J = data['J']
    return alpha, theta, J

# Read the data
file_path = 'data/3D_data-tpssh.csv'
alpha, theta, J = read_data_from_csv(file_path)

# Create a 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with data points using the same colormap as density map
sc = ax.scatter(theta, alpha, J, c=J, cmap='coolwarm', marker='o', edgecolor='red')

# Set labels for each axis
ax.set_xlabel('Theta', weight='bold')
ax.set_ylabel('Alpha', weight='bold')
ax.set_zlabel('J', weight='bold')

# Create a continuous surface for the density map
triang = mtri.Triangulation(theta, alpha)
ax.plot_trisurf(triang, J, cmap='coolwarm', edgecolor='none', alpha=0.5)

# Add a color bar to show the scale of J values
cbar = plt.colorbar(sc, location='right', pad=0.1, ax=ax)
# cbar.set_label('J', weight='bold')

# Show the plot
plt.show()
```
